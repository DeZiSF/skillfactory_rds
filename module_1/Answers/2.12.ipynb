```python
import pandas as pd
df = pd.read_csv('data_sf.csv')
```

```python
pivot = df.pivot_table(
    values=['Wage'],
    index=['Nationality'],
    columns=['Club'],
    aggfunc='sum',
    fill_value = 0
)
display(pivot)
```

```python
pivot = df.pivot_table(
    values = ['Name'],
    index = ['Position'],
    columns = ['Club'],    
    aggfunc = 'count',
    fill_value = 0
)

display(pivot.loc['CM'].value_counts())
```

```python
pivot = df.pivot_table(
    values = ['Wage'],
    index = ["Nationality"],
    columns = ["Club"],
    aggfunc = 'sum'    
)

display(pivot.loc['Russia']['Wage']['AS Monaco'])
```

```python
g = df[df.Position == 'ST'].groupby(['Club'])['SprintSpeed'].mean().sort_values(ascending = False)
display(g)
```
