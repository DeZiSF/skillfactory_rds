```python
def profession_code(str):
    if str == 'Рабочий':
        return 0
    if str == 'Менеджер':
        return 1
    return 2
```

```python
sample = pd.read_csv("sample.csv")
sample.Profession = sample.Profession.apply(profession_code)
sample2 = sample

sample2.head()
```

```python
def age_category(age):
    if age < 23:
        return 'молодой'
    if age < 35:
        return 'средний'
    return 'зрелый'
```

```python
sample = pd.read_csv("sample.csv")
sample['Age_category'] = sample.Age.apply(age_category)
sample.head()
```

```python
def trim(name):
    return name.replace('Запись пользователя № - ', '').replace('#error', '')

log = pd.read_csv('log.csv', header=None)
log.columns = ['user_id','time','bet','win']

log.user_id = log.user_id.apply(trim)
log.head()
```
