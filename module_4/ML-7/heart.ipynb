```python
import pandas as pd
df = pd.read_csv('heart_fin.csv', ';', encoding = 'ISO-8859-1')
```

```python
def qr(s):
    iqr = s.quantile(0.75) - s.quantile(0.25)
    return s.mean() - 1.5*iqr, s.mean() + 1.5*iqr

def clear(data, column, qrp):
    return data[(data[column] >= qrp[0]) & (data[column] <= qrp[1])]

qr_age = qr(df['age'])
qr_cp = qr(df['cp'])
qr_trestbps = qr(df['trestbps'])
qr_chol = qr(df['chol'])
qr_restecg = qr(df['restecg'])
qr_thalach = qr(df['thalach'])


df = clear(df, 'age', qr_age)
df = clear(df, 'cp', qr_cp)
df = clear(df, 'trestbps', qr_trestbps)
df = clear(df, 'chol', qr_chol)
df = clear(df, 'restecg', qr_restecg)
df = clear(df, 'thalach', qr_thalach)
```
