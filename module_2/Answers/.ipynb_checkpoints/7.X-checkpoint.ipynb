```python
df['tip_pers'] = 100 * df.tip / df.total_bill
df['tip_pers'].plot(kind = 'hist')
```

```python
df.groupby(by = ['sex', 'smoker'])['tip'].mean().plot(kind = 'bar', grid= True)
```

```python
df.plot(x='total_bill', y = 'tip_pers', kind = 'scatter')
```
