```python
with open('recipes.json') as f:
    recipes = json.load(f)

df = pd.DataFrame(recipes)

def find_item(item, cell):
    if item in cell:
        return 1
    return 0

ingredients = set()
def add_ingrid(cell):
    for i in cell:        
        ingredients.add(i)

df['ingredients'].apply(lambda cell: add_ingrid(cell))

for i in ingredients:
    df[i] = df['ingredients'].apply(lambda cell: find_item(i, cell))

df['ingredients'] = df['ingredients'].apply(lambda cell: len(cell))
```
