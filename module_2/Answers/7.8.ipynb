```python
import numpy as np
x = np.linspace(start = -3., stop = 3., num = 1000)
y1 = np.exp(x)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fig = plt.figure()
axes = fig.add_axes([0, 0, 1, 1])
axes.plot(x, y1, linestyle='dashed')
```
