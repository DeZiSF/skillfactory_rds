```python
balance = df['balance_due']

def iqr_f(ys):
    q25, q75 = np.percentile(ys, [25, 75])
    iqr = q75 - q25
    lower = q25 - 1.5 * iqr
    upper = q75 + 1.5 * iqr
    return np.where((ys < lower) | (ys > upper))[0]

o = iqr_f(balance.dropna())
o = balance.iloc[o]
mx = max(o)
mn = min(o)
display(o)
display(abs(mx - mn))
```
